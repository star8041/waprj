## 결과 불러오기

```python
import json
import pandas as pd

with open("final_results.json", "r", encoding="utf-8") as f:
    final_results = json.load(f)

df = pd.DataFrame(final_results)
print(f"✅ {len(df)}건 로드 완료")
df
```

## csv 저장

```python
import os
os.makedirs("results", exist_ok=True)

# 필요한 컬럼만 선택해서 저장
columns = ["제목", "발신자", "분류_전체", "신뢰도_전체", "액션필요", "기한", "요약"]
df[columns].to_csv("results/mail_classify_result.csv", index=False, encoding="utf-8-sig")
print("✅ CSV 저장 완료 → results/mail_classify_result.csv")
```

## 최종 요약 출력

```python
print("=" * 40)
print("최종 분류 결과 요약")
print("=" * 40)
print(df["분류_전체"].value_counts().to_string())
print()
print(f"전체 메일 수             : {len(df)}건")
print(f"액션 필요 메일           : {df['액션필요'].sum()}건")
print(f"기한 있는 메일           : {(df['기한'] != '-').sum()}건")
print(f"평균 신뢰도 (제목)       : {df['신뢰도_제목'].mean():.2%}")
print(f"평균 신뢰도 (제목+본문)  : {df['신뢰도_제목+본문'].mean():.2%}")
print(f"평균 신뢰도 (전체)       : {df['신뢰도_전체'].mean():.2%}")
```

## 카테고리별 PDF 폴더 확인

```python
print("=" * 40)
print("카테고리별 PDF 현황")
print("=" * 40)

categories = ["업무협조", "보고서", "회의록", "공지", "기타"]
for cat in categories:
    folder = os.path.join("downloads", cat)
    if os.path.exists(folder):
        files = [f for f in os.listdir(folder) if f.endswith(".pdf")]
        print(f"{cat:10} : {len(files)}개")
    else:
        print(f"{cat:10} : 폴더 없음")
```

## 전체 파일 구조 출력


```python
print("=" * 40)
print("downloads 폴더 구조")
print("=" * 40)

for cat in categories:
    folder = os.path.join("downloads", cat)
    if not os.path.exists(folder):
        continue
    files = [f for f in os.listdir(folder) if f.endswith(".pdf")]
    print(f"\n📁 {cat}/")
    for f in files:
        print(f"   └── {f}")
```
