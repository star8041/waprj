# IMAP 연결

## 계정 입력

```python
EMAIL = "아이디"
PASSWORD = "앱_비밀번호"  # 앱 비밀번호 16자리
```

## IMAP 접속

```python
import imaplib

mail = imaplib.IMAP4_SSL("imap.gmail.com")
mail.login(EMAIL, PASSWORD)

print("접속 성공")
```

## 메일 개수 확인

```python
mail.select("inbox")

_, msg_nums = mail.search(None, "ALL")
total = len(msg_nums[0].split())

print(f"받은편지함 메일 수: {total}건")
```

## 접속 종료

```python
mail.logout()
print("접속 종료")
```
